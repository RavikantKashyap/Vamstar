# Objective:

To create Python BMI Calculator

# Import Libraries

```python
import pandas as pd 
```

# Provided Json data

```python
dict1 = [{"Gender": "Male", "HeightCm": 171, "WeightKg": 96},
{ "Gender": "Male", "HeightCm": 161, "WeightKg": 85},
{ "Gender": "Male", "HeightCm": 180, "WeightKg": 77},
{ "Gender": "Female", "HeightCm": 166, "WeightKg": 62},
{"Gender": "Female", "HeightCm": 150, "WeightKg": 70},
{"Gender": "Female", "HeightCm": 167, "WeightKg": 82}]
```

# Converting into dataframe and initiated the tasks

```python
df = pd.DataFrame(dict1)
df 
```

```python
df['Height_meter'] = df['HeightCm']/100
df 
```

```python
df['Height_meter_square']= df['Height_meter']**2
df 
```

# Calculating the BMI

```python
df['BMI'] = df['WeightKg']/df['Height_meter_square']
df 
```

# Creating BMI Category column 

```python
def BMI_category_fun(BMI_value):
    temp=[]
    if BMI_value<=18.4:
        temp.append('Underweight')
        
    elif (BMI_value>=18.5) &(BMI_value<=24.9):
        temp.append('Normal weight')
       
    elif (BMI_value>=25) &(BMI_value<=29.9):
        temp.append('Overweight')
        
    elif (BMI_value>=30) &(BMI_value<=34.9):
        temp.append('Moderately obese')
       
    elif (BMI_value>=35) &(BMI_value<=39.9):
        temp.append('Severely obese')
        
    elif BMI_value>=40:
        temp.append('Very severely obese ')

    return temp
```

```python
df['BMI_Cat'] = df['BMI'].apply(BMI_category_fun)
```

```python
df['BMI Category'] = [' '.join(map(str, l)) for l in df['BMI_Cat']]
```

```python
df = df.drop(columns={'BMI_Cat'})
df 
```

# Creating Health risk column 

```python
def BMI_health_fun(BMI_value):
    temp=[]
    if BMI_value<=18.4:
        temp.append('Malnutrition risk')
        
    elif (BMI_value>=18.5) &(BMI_value<=24.9):
        temp.append('Low risk')
       
    elif (BMI_value>=25) &(BMI_value<=29.9):
        temp.append('Enhanced risk')
        
    elif (BMI_value>=30) &(BMI_value<=34.9):
        temp.append('Medium risk')
       
    elif (BMI_value>=35) &(BMI_value<=39.9):
        temp.append('High risk')
        
    elif BMI_value>=40:
        temp.append('Very high risk')

    return temp
```

```python
df['BMI_health'] = df['BMI'].apply(BMI_health_fun)
```

```python
df['Health risk'] = [' '.join(map(str, l)) for l in df['BMI_health']]
```

```python
df = df.drop(columns={'BMI_health'})
df 
```

# Count the total number of overweight people using ranges in the column BMI Category


```python
tot_num_of_overweight_people = df.loc[df['BMI Category']=='Overweight']
tot_num_of_overweight_people
```
